```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
```

```python
#data
df=pd.read_csv('Crimes_-_2001_to_Present (1).csv', low_memory=False)
```

```python
# Calculate the overall arrest rate
total_arrests = df['Arrest'].sum()
total_crimes = len(df)
arrest_rate = total_arrests / total_crimes
print(f"Overall Arrest Rate: {arrest_rate:.2%}")
```

```python

# Calculate the arrest rate by primary type
arrest_rates = df.groupby('Primary Type')['Arrest'].mean().sort_values()
ax = arrest_rates.plot(kind='barh', figsize=(10,10))
ax.set_title('Arrest Rate by Primary Type')
ax.set_xlabel('Arrest Rate')
ax.set_ylabel('Primary Type')
plt.savefig('Arrest_Rate_Type.png', bbox_inches='tight')
plt.show()
```

```python

# Calculate the arrest rate by police district
arrest_rates = df.groupby('District')['Arrest'].mean().sort_values()
ax = arrest_rates.plot(kind='barh', figsize=(10,10))
ax.set_title('Arrest Rate by Police District')
ax.set_xlabel('Arrest Rate')
ax.set_ylabel('Police District')
plt.savefig('Arrest_Rate_District.png', bbox_inches='tight')
plt.show()
```
